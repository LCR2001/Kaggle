 : 결정 경계를 정의하는 것이 결국 서포트 벡터임 
  - 데이터 중에서 서포트 벡터만 잘 골라낼 경우 다른 데이터들은 무시 가능

# MODEL TRAINING ( FINDING A PROBLEM SOLUTION )

```python
# 입력값과 출력값 정의
# target 삭제 후 axis =1 -> 열 지정
X = df_cancer.drop(['target'], axis = 1)
```

```python
X
```

```python
y = df_cancer['target']
y
```

```python
from sklearn.model_selection import train_test_split
# random_state : random 함수의 seed값과 유사
# -> 값 지정 안 해줄 경우 코드 실행할 때마다 값이 다르게 나옴.(학습/데이터 세트가 다르게 지정됨)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 5)
```

```python
X_train.shape 
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# SVM : 서포트 벡터 머신 활용
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

svc_model = SVC()
svc_model.fit(X_train, y_train)
```

# EVALUATING THE MODEL
- confusion matrix ( 오차행렬 )
 : 학습 혹은 테스트의 결과를 보여주는 행렬
  - 행 : Predict
  - 열 : True class
    - 올바른 예측 : TN (True Negative) + TP (True Positive)
    - 오분류   : FN (False Negative) + FP (False Positive)
  - 정밀도 (Precision) : 1로 예측했을 때 얼마나 잘 맞췄는지
    - TP / (FP + TP)
  - 재현율 (Recall) : 실제 1인 값을 얼마나 잘 찾았는지
    - TP / (FN + TP)
  - 1종 오류 (Type 1 Error) : 예측을 1로 했지만 실제는 0인 경우
  - 2종 오류 (Type 2 Error) : 예측을 0이라고 했지만 실제는 1이 경우 --> 위험이 더 큼!

```python
y_predict = svc_model.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True)
```

- TN : 0   /  FN : 7
- FP : 0   /  TP : 66

```python
print(classification_report(y_test, y_predict))
```

# IMPROVING THE MODEL 1
: 데이터 범위를 0~1로 축소 ( Data Normalization) _ Feature scaling Unity Normalization

1. 데이터 정규화(Data Normalization)
- Feature scaling : 통일성 기반 정규화(Unity-based normalization) 사용 -> 0~1 사이의 값으로 데이터를 만듦.
    - (X-X_min) / (X_max - X_min)

2. SVM 파라미터 조정(SVM Parameters Optimization)
- C parameter : 학습 지점의 올바른 분류와 원만한 결정경계 사이의 균형을 제어 -> 모델이 데이터 지점을 오분류할 시 패널티(cost) 적용
    - small C : 오분류에 대해 cost를 적게 부여. soft margin_두 개 클래스 사이에 더 평탄한 결정경계를 얻을 수 있음.
    - large C : 오분류에 대해 cost를 크게. hard margin_두 개 클래스 사이에 복잡한(더 엄격하거나 더 잘 맞는) 경계를 얻을 수 있음. 
        - 학습 데이터 세트에만 특정 ->  과적합에 취약함.
    
- Gamma parameter : 단일 학습 세트의 영향이 미치는 범위를 제어함(데이터 포인트의 영향 범위를 지정)
    - small Gamma : 범위가 넓음. 더 일반화된 솔루션을 가짐.
        - 결정경계만 고려하는 것이 아닌, 훨씬 더 많은 지점(공간)을 고려.
    - large Gamma : 범위가 짧음_초평면(결정경계)에 가까운 값에만 집중
        - 과적화될 위험 높음 <- 초평면에 멀리 떨어진 다른 지점은 무시

- C, Gamma parameter 값 조정 방법: 그리드 검색(Grid search)

```python
min_train = X_train.min()
min_train
```

```python
range_train = (X_train - min_train).max()
range_train
```

```python
X_train_scaled = (X_train - min_train) / range_train
```

```python
X_train_scaled
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test) / range_test
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

svc_model = SVC()
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_predict)

sns.heatmap(cm, annot = True, fmt = 'd')
```

```python
print(classification_report(y_test, y_predict))
```

# IMPROVING THE MODEL 2
: SVC의 Parameter값 조정 (C, Gamma)

```python
# 커널 rbf : 방사형 기저 함수
# kernel -> linear : C parameter만 사용 / rbf : C, Gamma parameter 둘 다 사용
param_grid = {'C' : [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# refit : True면 가장 최적의 하이퍼 파라미터를 찾은 뒤 입력된 estimator 객체를 해당 하이퍼 파라미터로 재학습시킨다
# verbose : GridSearchCV의 iteration시마다 검색 도중 표시할 값의 수
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test_scaled)
cm = confusion_matrix(y_test, grid_predictions)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,grid_predictions))
```
